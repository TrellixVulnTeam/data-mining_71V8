```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn as skl
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from sklearn.datasets import load_iris
from sklearn import metrics, model_selection, linear_model
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse
```

```python
data = load_iris()
```

```python
dfRaw = pd.DataFrame(data= np.c_[data['data'], data['target']],
                     columns= data['feature_names'] + ['target'])

dfRaw
```

```python
df = dfRaw # Create working copy.
x = df.drop(['target'], axis=1)
y = df["target"]
```

```python
def GetClassName(data, n):
    n = int(n)
    return data.target_names[n]

def PlotXY(x, y):
    plt.figure(figsize=(20,10))
    nCols = len(x.columns.values)
    i =0
    for col in x.columns.values:
        plt.subplot(2, nCols, i + 1 )
        plt.xlabel(col)
        plt.ylabel("Class")
        plt.scatter(x[col], y, marker="x")
        i += 1
```

```python
PlotXY(x,y)
```

```python
xtr, xte, ytr, yte = skl.model_selection.train_test_split(x, y, test_size=0.3, shuffle=True)
```

```python
PlotXY(xtr, ytr)
```

```python
PlotXY(xte, yte)
```
